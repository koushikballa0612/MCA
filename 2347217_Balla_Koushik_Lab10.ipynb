1. Create a form using Tkinter on your Domain details. (Domain name, Student name and other
details, with text boxes, radio buttons, menu and the required widgets you can use)

```python
import tkinter as tk
from tkinter import ttk

def submit():
    student_name = student_name_entry.get()
    dob = dob_entry.get()
    selected_option = gender_var.get()
    selected_country = "India"
    selected_state = state_var.get()
    pincode = pincode_entry.get()
    selected_course = course_var.get()
    selected_payment_option = payment_var.get()
    
    # You can process the data here, e.g., print it to the console
    print("Course Selection and Payment Page")
    print(f"Student Name: {student_name}")
    print(f"Date of Birth: {dob}")
    print(f"Gender: {selected_option}")
    print(f"Country: {selected_country}")
    print(f"State: {selected_state}")
    print(f"Pincode: {pincode}")
    print(f"Course: {selected_course}")
    print(f"Payment Option: {selected_payment_option}")

# Function to change the background color when the mouse hovers over an option
def on_enter(event):
    event.widget.itemconfig(tk.ACTIVE, {'background': 'lightgray'})

# Function to reset the background color when the mouse leaves an option
def on_leave(event):
    event.widget.itemconfig(tk.ACTIVE, {'background': 'white'})

# Create the main window
root = tk.Tk()
root.title("Domain Details Form")

# Display the domain name as a label
domain_name_label = ttk.Label(root, text="Course Selection and Payment Page")
domain_name_label.grid(row=0, column=0, columnspan=4, padx=10, pady=10, sticky='w')

# Create labels and entry widgets for Student Name and DOB
student_name_label = ttk.Label(root, text="Student Name:")
student_name_label.grid(row=1, column=0, padx=10, pady=10, sticky='w')
student_name_entry = ttk.Entry(root)
student_name_entry.grid(row=1, column=1, padx=10, pady=10, sticky='w')

dob_label = ttk.Label(root, text="Date of Birth (YYYY-MM-DD):")
dob_label.grid(row=2, column=0, padx=10, pady=10, sticky='w')
dob_entry = ttk.Entry(root)
dob_entry.grid(row=2, column=1, padx=10, pady=10, sticky='w')

# Create a radio button group for gender
gender_label = ttk.Label(root, text="Gender:")
gender_label.grid(row=3, column=0, padx=10, pady=10, sticky='w')
gender_var = tk.StringVar()
male_radio = ttk.Radiobutton(root, text="Male", variable=gender_var, value="Male")
female_radio = ttk.Radiobutton(root, text="Female", variable=gender_var, value="Female")
male_radio.grid(row=3, column=1, padx=10, pady=10, sticky='w')
female_radio.grid(row=3, column=2, padx=10, pady=10, sticky='w')

# Create labels and entry widgets for Address, State, and Pincode
address_label = ttk.Label(root, text="Address:")
address_label.grid(row=4, column=0, padx=10, pady=10, sticky='w')
address_entry = ttk.Entry(root)
address_entry.grid(row=4, column=1, padx=10, pady=10, sticky='w')

state_label = ttk.Label(root, text="State in India:")
state_label.grid(row=5, column=0, padx=10, pady=10, sticky='w')
states_in_india = ["Andhra Pradesh", "Arunachal Pradesh", "Assam", "Bihar", "Chhattisgarh", "Goa", "Gujarat", "Haryana", "Himachal Pradesh", "Jharkhand", "Karnataka", "Kerala", "Madhya Pradesh", "Maharashtra", "Manipur", "Meghalaya", "Mizoram", "Nagaland", "Odisha", "Punjab", "Rajasthan", "Sikkim", "Tamil Nadu", "Telangana", "Tripura", "Uttar Pradesh", "Uttarakhand", "West Bengal"]
state_var = tk.StringVar()
state_dropdown = ttk.Combobox(root, textvariable=state_var, values=states_in_india)
state_dropdown.grid(row=5, column=1, padx=10, pady=10, sticky='w')

# Bind mouse enter and leave events to the Combobox
state_dropdown.bind('<Enter>', on_enter)
state_dropdown.bind('<Leave>', on_leave)

pincode_label = ttk.Label(root, text="Pincode:")
pincode_label.grid(row=6, column=0, padx=10, pady=10, sticky='w')
pincode_entry = ttk.Entry(root)
pincode_entry.grid(row=6, column=1, padx=10, pady=10, sticky='w')

# Create a dropdown menu for selecting the course
course_label = ttk.Label(root, text="Course:")
course_label.grid(row=7, column=0, padx=10, pady=10, sticky='w')
courses = ["Mathematics", "Science", "History", "Computer Science", "Other"]
course_var = tk.StringVar()
course_dropdown = ttk.Combobox(root, textvariable=course_var, values=courses)
course_dropdown.grid(row=7, column=1, padx=10, pady=10, sticky='w')

# Create a label and radio buttons for payment options
payment_label = ttk.Label(root, text="Payment Option:")
payment_label.grid(row=8, column=0, padx=10, pady=10, sticky='w')
payment_var = tk.StringVar()
payment_var.set("Credit Card")  # Set a default value
credit_card_radio = ttk.Radiobutton(root, text="Credit Card", variable=payment_var, value="Credit Card")
debit_card_radio = ttk.Radiobutton(root, text="Debit Card", variable=payment_var, value="Debit Card")
net_banking_radio = ttk.Radiobutton(root, text="Net Banking", variable=payment_var, value="Net Banking")
credit_card_radio.grid(row=8, column=1, padx=10, pady=10, sticky='w')
debit_card_radio.grid(row=8, column=2, padx=10, pady=10, sticky='w')
net_banking_radio.grid(row=8, column=3, padx=10, pady=10, sticky='w')

# Create a submit button
submit_button = ttk.Button(root, text="Submit", command=submit)
submit_button.grid(row=9, column=0, columnspan=2, padx=10, pady=10)

root.mainloop()
```

2. Visualize your dataset using Matplotlib and write a program to create line graph, bar graph,
scatter plot and correlation graph or heat map.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module="seaborn")

# Load your dataset
data = pd.read_csv('udemy_courses.csv')  

# Convert 'published_timestamp' column to datetime
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'])

# Sort the DataFrame by 'published_timestamp' to ensure the data is in chronological order
data.sort_values(by='published_timestamp', inplace=True)

# Suppress warnings related to is_categorical_dtype
warnings.filterwarnings("ignore", category=UserWarning, module="seaborn")

# Line Graph
plt.figure(figsize=(10, 6))
plt.plot(data['published_timestamp'], data['num_subscribers'], marker='o', linestyle='-')
plt.title('Number of Subscribers Over Time')
plt.xlabel('Published Timestamp')
plt.ylabel('Number of Subscribers')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


# Bar Graph
plt.figure(figsize=(10, 6))
sns.barplot(x='level', y='num_lectures', data=data)
plt.title('Number of Lectures by Level')
plt.xlabel('Level')
plt.ylabel('Number of Lectures')
plt.tight_layout()
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(data['price'], data['num_reviews'], alpha=0.5)
plt.title('Price vs. Number of Reviews')
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.tight_layout()
plt.show()


# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
